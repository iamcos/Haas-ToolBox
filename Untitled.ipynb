```python

btc = pd.read_csv('/Users/cosmos/Documents/GitHub/Haas-ToolBox/MarketData/BTC_USDT_BINANCE_from_2011_03_01.csv')
eth = pd.read_csv('/Users/cosmos/Documents/GitHub/Haas-ToolBox/MarketData/ETH_USDT_BINANCE_from_2011_03_01.csv')
ltc = pd.read_csv('/Users/cosmos/Documents/GitHub/Haas-ToolBox/MarketData/LTC_USDT_BINANCE_from_2011_03_01.csv')

coins = {'btc':btc,'eth':eth,'ltc':ltc}
print(ltc.columns)
```

```python
# resampling  data
```

```python
for i in coins:
    coins[i]["Date"] = pd.to_datetime(coins[i]["Date"])
    dti = pd.DatetimeIndex([x for x in coins[i]["Date"]])
    coins[i].set_index(dti,inplace=True, drop=True)
    coins[i].drop(['Date',"Date.1",'Buy','Sell'],axis=1,inplace=True)
```

```python
for i in coins:
    print(i,coins[i])
```

```python
import plotly.express as px
```

```python
rs_coins = {}
for i in coins:
    rs = coins[i].resample('1W').mean()
    rs_coins[i] = rs

    print(i,rs_coins[i])
```

```python
pt_changes = {}
plt.close("all")
for i in rs_coins:
    pt_changes[i] = rs_coins[i].pct_change()

    
for i in pt_changes:
   
    
    fig = px.bar(pt_changes[i],x=pt_changes[i].index,y=pt_changes[i].Volume)
# print(pt_changes)
```
